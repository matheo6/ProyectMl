```python
import gzip
import json
import ast
import pandas as pd
```

```python
def readDataJson(ruta):
    data=[]
    with gzip.open(ruta, 'rb') as f:
        for line in f:
            data.append(json.loads(line.decode('utf-8')))
    return pd.DataFrame(data)
```

```python
def readDataAst(ruta):
    data=[]
    with gzip.open(ruta, 'rb') as f:
        for line in f:
            data.append(ast.literal_eval(line.decode('utf-8')))
    return pd.DataFrame(data)
```

```python
games2=readDataJson('steam_games.json.gz')
user_review2= readDataAst('user_reviews.json.gz')
user_items2=readDataAst('users_items.json.gz')
```

```python
games2.dropna(how='all',inplace=True)
user_review2.dropna(how='all',inplace=True)
user_items2.dropna(how='all',inplace=True)
```

```python
def unnest(df,column): 
    return pd.concat([df,pd.json_normalize(df.explode(column,ignore_index=True)[column])],axis=1).drop(columns=column)
```

```python
lista_reviews=unnest(user_review2,'reviews')
lista_items=unnest(user_items2,'items')
```
